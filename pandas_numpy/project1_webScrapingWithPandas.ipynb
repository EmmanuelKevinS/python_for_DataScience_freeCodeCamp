# Read .csv from Multiple URLs with Pandas



<b> Link: root + season + league </b>

```python
# link structure
"https://www.football-data.co.uk/mmz4281/" + "2425" + "/" + "E0" + ".csv"
```

```python
#Creating a root variable
root = 'https://www.football-data.co.uk/mmz4281/'
```

### Multiple leagues
https://www.football-data.co.uk/mmz4281/2425/E0.csv <br>
https://www.football-data.co.uk/mmz4281/2425/E1.csv <br>
https://www.football-data.co.uk/mmz4281/2425/E2.csv <br>
https://www.football-data.co.uk/mmz4281/2425/E3.csv <br>
https://www.football-data.co.uk/mmz4281/2425/EC.csv

```python
#creating a list of the leagues
leagues = ['e0','e1','e2','e3','ec']
frames = []

#looping through the leagues
for league in leagues:
    df = pd.read_csv(f'{root}/2425/{league}.csv')
    frames.append(df)
```

```python
#creating a list of the leagues
leagues = ['e0','e1','e2','e3','ec']
frames = []

#looping through the leagues
for league in leagues:
    for season in range(19, 25):
        df = pd.read_csv(f'{root}/{str(season)}{str(season+1)}/{league}.csv',encoding='unicode_escape')
        frames.append(df)
```

### Organize All The Data in a Dictionary

```python
#creating a dict with the league name as the key
dict_countries = {
    'Spanish La Liga':'SP1',
    'Spanish Segunda Division':'SP2',
    'German Bundesliga':'D1',
    'English Premier League':'E0',
    'English League 1':'E2',
    'English League 2':'E3'
}
```

```python
#getting a dictionary element
dict_countries['English League 2']
```

```python
dict_historical_data = {}


#looping through the leagues
for league in dict_countries:
    frames = []
    for season in range(23, 25):
        df = pd.read_csv(f'{root}/{str(season)}{str(season+1)}/{dict_countries[league]}.csv',encoding='unicode_escape')
        df.insert(1,'season',season)
        frames.append(df)
    df_concat = pd.concat(frames)
    dict_historical_data[league] = df_concat
```
