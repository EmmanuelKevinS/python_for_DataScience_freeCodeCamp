### adding new columns with insert() and assign()

```python
#just to keep remembering
import numpy as np
import random
```

```python
#creating random numbers for our new score columns
score1 = np.random.randint(0,101,size=1000)
score2 = np.random.randint(0,101,size=1000)
```

```python
#creating a series using the random numbers
series1 = pd.Series(score1, index=(np.arange(0,1000)))
series2 = pd.Series(score2,index=(np.arange(0,1000)))
```

```python
df_exams.assign(score1=series1,score2=series2)
```

```python
#trying a side quest
df_exams['parental level of education'].unique()

```

```python
score_ranges = {
    "master's degree": (75, 101),
    "bachelor's degree": (70, 97),
    "associate's degree": (65, 92),
    "some college": (60, 87),
    "high school": (45, 82),
    "some high school": (40, 77),
}
```

```python
def generate_language_scores(education):
    low, high = score_ranges[education]
    return np.random.randint(low,high)
```

```python
df_exams.head()
```

```python
df_exams['language score'] = df_exams['parental level of education'].apply(generate_language_scores)
```

```python
df_exams
```

```python
df_exams = df_exams.assign(score1=series1,score2=series2)
```

```python
df_exams = df_exams.drop(columns=['test'])
```

```python
#insert method to add a new column
# format: column number(you want to be inserted into,column name,data)
df_exams.insert(1,'test',series1)
```

```python
df_exams
```
